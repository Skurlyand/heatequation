```python
import matplotlib.pyplot as plt
import numpy as np
from math import*
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def f(n,m):
    if ((n%2)!=0) or (n==1):
        return 16/((pi**3)*(n**2)*(2*m+1))*(-1)**(m+1)
    else:
        if n==0:
            return 2*((-1)**(m))/pi/(2*m+1)
        else:
            return 0
def g(n, m):
        if n==0:
            return 32/(pi**3)/((2*m+1)**3)*(-1)**(m+1)
        else:
            return 0
        
def lam(m, n):
    return (pi*n)**2+(pi/2+pi*m)**2

def anal(n, m, t, y, x):
    return ((g(n,m)-2*f(n,m)/(lam(m,n)**3))*exp(-lam(m, n)*t)+f(n, m)*((lam(m,n)**2)*(t**2)-2*lam(m,n)*t+2)/(lam(m,n)**3))*cos(pi*n*y)*cos((pi/2+pi*m)*x)

N1, N2, J = 50, 50, 50 # step
X1, X2 = 0, 1 # region
Y1, Y2 = 0, 1
T1, T2 = 0, 1
to=int(0.8*J)
x = np.linspace(X1,X2,N1)
y = np.linspace(Y1,Y2,N2)
t = np.linspace(T1,T2,J)

h_1=x[1]-x[0]
h_2=y[1]-y[0]
tau=t[1]-t[0]

kx = tau / (h_1**2)
ky = tau / (h_2**2)

w=np.zeros((N1,N2,J*2+1))
wanal =np.zeros((N1,N2,J*2+1))

for n in range(0, 20): #analytical solution
    for m in range(0, 20):
        for j in range(0, N2):
            for k in range (0, N1):
                wanal[k, j, to] += anal(n, m, t[to], y[j], x[k])

#initial data
for i in range(0,N1):
    for j in range(0,N2):
        w[i,j,0]=(x[i])**2-1
        
def F1(i1, i2, j):
    return 0.5 * ky * (w[i1, i2 - 1, j - 1] + w[i1, i2 + 1, j - 1])+(1-ky)* w[i1, i2, j - 1] + 0.5 * tau * y[i2]*(tau*(j+1)*0.5)**2


def progx(i2, j):
    alpha = np.zeros(N1)
    beta = np.zeros(N1)
    alpha[0] = 1
    beta[0] = 0
    
    A = 0.5 * kx
    C = 1 + kx
    B = 0.5 * kx
    for m in range(1, N1-1): #straight prog
        Fm = F1(m, i2, j)
        alpha[m] = A / (C - B * alpha[m-1])
        beta[m] = ( Fm + B * beta[m-1]) / (C - B*alpha[m-1])
    beta[N1-1] = 0 #dirichle
    w[N1 - 1, i2, j] = beta[N1-1]
    for m in range(N1 - 1, 0, -1): #inverse prog
        w[m - 1, i2, j] = alpha[m-1] * w[m, i2, j] + beta[m-1]

def F2(i1, i2, j):
    return 0.5 * kx * (w[i1-1, i2, j - 1] + w[i1+1, i2, j - 1]) + (1 - kx)* w[i1, i2, j - 1] + 0.5 * tau * y[i2]*(tau*(j-1)*0.5)**2

def progy(i1, j):
    alpha = np.zeros(N2)
    beta = np.zeros(N2)
    alpha[0] = 1
    beta[0] = 0
    A = 0.5 * ky
    C = 1 + ky
    B = 0.5 * ky
    for m in range(1, N2-1): #straight prog
        Fm = F2(i1, m, j)
        alpha[m] = A / (C - B * alpha[m-1])
        beta[m] = ( Fm + B * beta[m-1]) / (C - B*alpha[m-1])
    beta[N2-1] = beta[N2-2]/(1-alpha[N2-2])#newman
    w[i1, N2-1, j] = beta[N2-1]
    for m in range(N2 - 1, 0, -1): #inverse prog
        w[i1, m-1, j] = alpha[m-1] * w[i1, m, j] + beta[m-1]
        
for j in range(1,2*J,2):
    for i2 in range(1,N2-1):
        progx(i2,j)
for i1 in range(0,N1-1):
    progy(i1,j+1)
    
u=np.zeros((N1,N2)) #graph of error
max = 0.
for j in range(0,N2):
    for i in range(0,N1):
        w[0, j, 2*to] = w[1, j, 2*to]
        u[i,j]=abs(w[i, j, 2*to]-wanal[i, j, to])
        if max < u[i, j]:
            max=u[i, j]

'''
u=np.zeros((N1,N2)) #graph of error
for j in range(0,N2):
for i in range(0,N1):
u[i,j]=wanal[i, j, 80]
'''
x, y = np.meshgrid(x, y)
ax = plt.axes(projection='3d')
ax.plot_surface(y, x, u, rstride=1, cstride=1, cmap = cm.jet)
plt.ylabel('y')
plt.xlabel('x')
plt.show()
print(max)
```
